```python
f = lambda t: np.exp(-t) * np.cos(2*np.pi*t)
a = np.arange(0, 5, 0.02, dtype=np.float)
```

*The codes have to group together .. kind of*

```python
plt.subplot(2, 1, 1) #2 row, 1 col, select row2col1
plt.plot(a, f(a))
plt.subplot(2, 1, 2) #2 row, 1 col, select row2col1
plt.plot(a, np.cos(2*np.pi*a), 'r--') #'r--' for dotted line
```

## More on Plot()

<br>plt.`plot`( x, y, fmt, **kwargs )<br>
<br> *fmt = '[color] [marker] [line]'*<br>
<br> *kwargs: more curves(x,y,fmt)*<br>

```python
a = np.arange(10)

plt.plot(a, a*1.5, a, a*2.5, a, a*3.5,\
        a, a*4.5)
```

**Color String**<br>
*<br> 'k', 'w' - black, white<br>
<br> 'r','b','g','y' ,'m', 'c' - red, blue, green, yellow, magenta, cyan <br>
<br> '0.x' - grey scale value<br>
<br> '#xxxxxx' - RGB value*

**Marker String**<br>
'.'	point marker <br>
','	pixel marker<br>
'o'	circle marker<br>
'v'	triangle_down marker<br>
'^'	triangle_up marker<br>
'<'	triangle_left marker<br>
'>'	triangle_right marker<br>

**Style String**<br>
*<br> '-' : ____(default)<br>
<br> '--' : ---- <br>
<br> '-.'  -.-. <br>
<br> ':'  ....<br>
<br> ''''  None<br>*

```python
#Play With Me
plt.title('Play With Format')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(a, a*1.5, 'r.-',\
         a, a*2.5, 'go--',\
         a, a*3.5, 'b<-.',\
         a, a*4.5, 'cD:',\
         a, a*5.5, 'y+'''\
         )
#a, a*6.5, color='#123456',marker = 'h', linestyle = 'dashed'
```

### Font Modification<br>
<br>*Mandatory for Chinese characters*

```python
#Avaliable fonts
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
#Global Change
matplotlib.rcParams['font.family']='Helvetica'
matplotlib.rcParams['font.style']='italic'
matplotlib.rcParams['font.size']=20
```

```python
#Change back to Default
matplotlib.style.use('default')
```

```python
#Local Change (recommend)
plt.title('Play With Font'\
          ,fontproperties='Arial', fontsize=20)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(a, a*1.5, 'r.-',\
         a, a*2.5, 'go--',\
         a, a*3.5, 'b<-.',\
         a, a*4.5, 'cD:',\
         a, a*5.5, 'y+'''\
         )
```

### Text Display

<br>plt.`text`(xpos, ypso, txt)

<br>plt.`annotate`( txt, xy= arrow_crd, xytext= text_text_crd, arrowprops= dict )

```python
swave = lambda x: np.cos(2*np.pi*x)
x = np.arange(0, 5, 0.02, dtype=np.float)
```

```python
plt.plot(x, swave(x))
plt.title(r'Sin Wave $y=cos(2\pi x)$'\
          ,fontproperties='Arial', fontsize=20)
#$ $Latex
plt.xlabel('x axis',fontsize=10 )
plt.ylabel('y axis',fontsize=10)
plt.text(5, 1, r'$\mu=100$') #(xpos, ypos, text)
plt.annotate('THIS',xy=(2,1), xytext=(3,1.5),\
            arrowprops=dict(facecolor='red',shrink=0.1,width=2))
plt.axis([-1, 6, -2, 2])
plt.grid(True)
```
