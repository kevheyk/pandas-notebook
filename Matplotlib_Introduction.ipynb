## More on Plot()

```python
a = np.arange(10)

plt.plot(a, a*1.5, a, a*2.5, a, a*3.5,\
        a, a*4.5)
```

**Color String**<br>

<br> 'r','b','g','y' ,'m', 'c' - red, blue, green, yellow, magenta, cyan <br>
<br> '0.x' - grey scale value<br>
<br> '#xxxxxx' - RGB value*

**Marker String**<br>
'.'	point marker <br>
','	pixel marker<br>
'o'	circle marker<br>
'v'	triangle_down marker<br>
'^'	triangle_up marker<br>
'<'	triangle_left marker<br>
'>'	triangle_right marker<br>

**Style String**<br>
*<br> '-' : ____(default)<br>
<br> '--' : ---- <br>
<br> '-.'  -.-. <br>
<br> ':'  ....<br>
<br> ''''  None<br>*

```python
#Play With Me
plt.title('Play With Format')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(a, a*1.5, 'r.-',\
         a, a*2.5, 'go--',\
         a, a*3.5, 'b<-.',\
         a, a*4.5, 'cD:',\
         a, a*5.5, 'y+'''\
         )
#a, a*6.5, color='#123456',marker = 'h', linestyle = 'dashed'
```

## Ploar Plot

[Polar Docs](https://matplotlib.org/gallery/misc/transoffset.html#sphx-glr-gallery-misc-transoffset-py)

```python
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import numpy as np
```

```python
xs = np.arange(7)
ys = xs**2
```

```python
fig = plt.figure(figsize=(5, 10))
ax = plt.subplot(2, 1, 1)
```

```python
# If we want the same offset for each text instance,
# we only need to make one transform.  To get the
# transform argument to offset_copy, we need to make the axes
# first; the subplot command above is one way to do this.
trans_offset = mtransforms.offset_copy(ax.transData, fig=fig,\
                                       x=0.05, y=0.10, units='inches')
```

```python
for x, y in zip(xs, ys):
    plt.plot((x,), (y,), 'ro')
    plt.text(x, y, '%d, %d' % (int(x), int(y)), transform=trans_offset)
```

```python
# offset_copy works for polar plots also.
ax = plt.subplot(2, 1, 2, projection='polar')
```

```python
trans_offset = mtransforms.offset_copy(ax.transData, fig=fig,
                                       y=6, units='dots')
```

```python
for x, y in zip(xs, ys):
    plt.polar((x,), (y,), 'ro')
    plt.text(x, y, '%d, %d' % (int(x), int(y)),
             transform=trans_offset,
             horizontalalignment='center',
             verticalalignment='bottom')
plt.show()
```
